```python
import numpy as np

print('Total number of Tracks : {}'.format(len(google_listened_to_df)))

google_listened_to_df



display(google_listened_to_df.head())
google_listened_to_df = google_listened_to_df.sort_values('Play Count', ascending=False)
display(google_listened_to_df.head())
google_listened_to_df.to_csv('C:\\Users\\alan__000\\PycharmProjects\\GoogleMusicDataAnalysis\\takeout_data\\Google Play Music\\combined_tracks2.csv')

# cleaned_df = google_listened_to_df.filter(google_listened_to_df['Play Count'] > 10)

# len(cleaned_df)

# combined_df.sort_values('Duration (ms)', ascending=False)
print(google_listened_to_df['Duration (ms)'].mean())
# remove outliers
combined_outliers_trimmed_df = google_listened_to_df[np.abs(google_listened_to_df['Duration (ms)']-google_listened_to_df['Duration (ms)'].mean()) <= (3*google_listened_to_df['Duration (ms)'].std())]
combined_outliers_trimmed_df.sort_values('Duration (ms)', ascending=False)

print(combined_outliers_trimmed_df['Duration (ms)'].mean())
```
