```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import pandas as pd
import seaborn as sns
import numpy as np

google_listened_to_df = pd.read_csv('C:\\Users\\alan__000\\PycharmProjects\\GoogleMusicDataAnalysis\\takeout_data\\Google Play Music\\combined_tracks.csv')
google_listened_to_df = google_listened_to_df.sort_values('Play Count', ascending=False)

combined_outliers_trimmed_df = google_listened_to_df[np.abs(google_listened_to_df['Duration (ms)']-google_listened_to_df['Duration (ms)'].mean()) <= (3*google_listened_to_df['Duration (ms)'].std())]
combined_outliers_trimmed_df.sort_values('Duration (ms)', ascending=False)

display(sns.distplot(combined_outliers_trimmed_df['Duration (ms)']))
```

```python
# take the jumble opf tracks and group this up by Artist
grouped_df = google_listened_to_df.groupby(['Artist'], as_index=False)['Play Count'].sum()
grouped_df = grouped_df.sort_values('Play Count', ascending=False)
display(grouped_df.head())

# lets index this so we can properly plot it etc
grouped_indexed_df = google_listened_to_df.groupby(['Artist'], as_index=True)['Play Count'].sum()
grouped_indexed_df = grouped_indexed_df.sort_values(ascending=False)
```
